```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Uploading our dataset
df = pd.read_excel("C:/Users/Faith/Downloads/ehealth-kenya-facilities.xlsx")
df.head(5)
```

```python
#Checking for the data types
df.dtypes
```

```python
#Summary statistics
df.describe().T
```

```python
#Checking for missing values in our dataset
df.isnull().sum()
```

```python
# List of columns to drop
columns_to_drop = ['Post Code','ANC', 'ART', 'BEOC', 'BLOOD', 'CAES SEC', 'CEOC', 'C-IMCI', 
                   'EPI', 'FP', 'GROWM', 'HBC', 'HCT', 'IPD', 'OPD', 'OUTREACH', 
                   'PMTCT', 'RAD/XRAY', 'RHTC/RHDC', 'TB DIAG', 'TB LABS', 
                   'TB TREAT', 'YOUTH', 'Official Landline', 'Official Fax', 
                   'Official Mobile', 'Official Email', 'Official Alternate No']

# Drop the columns
df.drop(columns=columns_to_drop, inplace=True)

# Check the resulting DataFrame
df.head(5)
```

```python
# Replace missing values with 0
df['Beds'] = df['Beds'].fillna(0)
df['Cots'] = df['Cots'].fillna(0)
```

```python
# Impute missing values with specific category
specific_category = 'NaN'

df['Division'] = df['Division'].fillna(specific_category)
df['Location'] = df['Location'].fillna(specific_category)
df['Sub Location'] = df['Sub Location'].fillna(specific_category)
df['Description of Location'] = df['Description of Location'].fillna(specific_category)
```

```python
# Defining replacement values for each column
replacement_values = {
    'Open 24 Hours': 'N/A',
    'Open Weekends': 'N/A',
    'Operational Status': 'Unknown'
}

# Replace missing values based on the defined replacement values
for column, value in replacement_values.items():
    df[column] = df[column].fillna(value)
```

```python
# Define replacement value
replacement_value = "Unknown"

# Replace missing values in the specified columns
columns_to_replace = ['Nearest Town', 'Official Address', 'Town', 'In Charge', 'Job Title of in Charge']

for column in columns_to_replace:
    df[column] = df[column].fillna(replacement_value)
```

```python
#Now checking on missing values for our final dataset after replacing and dropping the missing values
df.isnull().sum()
```

```python
# Save the DataFrame to a CSV file
df.to_csv('health_data.csv', index=False)
print("Dataset saved as 'health_data.csv'")
```

# Exploratory Data Analysis

We will answer a few questions based on certain topics in order to understand better our dataset

a) Healthcare Accessibility

What is the distribution of healthcare facilities across different regions and neighborhoods within Nairobi County?

```python
#We filter the dataset to include only healthcare facilities within Nairobi County
nrb_health_facilities_df = df[df['County'] == 'Nairobi']

# Group the healthcare facilities by regions and count the number of facilities in each region
facility_distribution_by_region = nrb_health_facilities_df.groupby('Constituency')['Facility Name'].count().reset_index()

# Sort the distribution by the number of facilities in descending order
facility_distribution_by_region = facility_distribution_by_region.sort_values(by='Facility Name', ascending=False)

# Display the distribution of healthcare facilities across different regions in Nairobi County
print("Distribution of Healthcare Facilities Across Different Constituencies in Nairobi County:")
print(facility_distribution_by_region)
```

```python
# Set the plot style
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Facility Name', y='Constituency', data=facility_distribution_by_region, palette='viridis')

# Add labels and title
plt.xlabel('Number of Facilities')
plt.ylabel('Constituency')
plt.title('Distribution of Healthcare Facilities Across Different Constituencies in Nairobi County')

# Show the plot
plt.show()
```

Which regions in Nairobi have the highest and lowest density of healthcare facilities per capita?

```python
#Upload our dataset
df3 = pd.read_excel("C:/Users/Faith/Downloads/kenya-population-by-sub-county (1).xlsx")
df3.head(5)
```

```python
# Merge the healthcare facilities dataset with the demographics dataset based on the common region/area
merged_df = pd.merge(df, df3, on='Sub Location')

# Calculate the total number of healthcare facilities in each area
facility_count_by_area = merged_df.groupby('Sub Location')['Facility Name'].count().reset_index()

# Calculate the population of each area
population_by_area = merged_df.groupby('Sub Location')['Total'].sum().reset_index()

# Merge the facility count and population dataframes
density_df = pd.merge(facility_count_by_area, population_by_area, on='Sub Location')

# Filter data for Subcounties in Nairobi
merged_df = merged_df[merged_df['County'] == 'Nairobi']

# Calculate the density of healthcare facilities per capita for each area
density_df['Density per Capita'] = density_df['Facility Name'] / density_df['Total']

# Identify the area with the highest density
highest_density_area = density_df.loc[density_df['Density per Capita'].idxmax()]

# Identify the area with the lowest density
lowest_density_area = density_df.loc[density_df['Density per Capita'].idxmin()]

print("Area with the Highest Density of Healthcare Facilities per Capita:")
print(highest_density_area)

print("\nArea with the Lowest Density of Healthcare Facilities per Capita:")
print(lowest_density_area)
```

Are there any underserved or remote areas within Nairobi that lack adequate access to healthcare facilities?

```python
#Defining the threshold for underserved areas
threshold = 5 

#Filter the dataset to include only Nairobi
nrb_health_facilities_df = df[df['County'] == 'Nairobi']

# Group the data by neighborhoods or regions and count the number of healthcare facilities
facility_count_by_area = nrb_health_facilities_df.groupby('Nearest Town')['Facility Name'].count().reset_index()

# Identify underserved areas with a low number of healthcare facilities
underserved_areas = facility_count_by_area[facility_count_by_area['Facility Name'] < threshold]

# Print the underserved areas
if not underserved_areas.empty:
    print("Underserved or remote areas within Nairobi lacking adequate access to healthcare facilities:")
    print(underserved_areas)
else:
    print("There are no underserved or remote areas within Nairobi lacking adequate access to healthcare facilities.")
```

b) Healthcare System Improvements

What are the primary healthcare needs and priorities identified by residents of Nairobi County?

```python
#We filter the dataset to include only Nairobi County
nrb_healthcare_priorities = df[df['County'] == 'Nairobi']

#Then we count the frequency of responses for types of healthcare facilities and priorities
healthcare_priorities_counts = nrb_healthcare_priorities['Type'].value_counts()

print("Frequency of responses for types of healthcare facilities and priorities:")
print(healthcare_priorities_counts)
```

```python
# Create the plot
plt.figure(figsize=(10, 6))
healthcare_priorities_counts.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel('Type of Healthcare Facility or Priority')
plt.ylabel('Frequency')
plt.title('Types of Healthcare Facilities and Priorities in Nairobi County')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')
```

How does the quality of healthcare services provided by different facilities vary across Nairobi?

```python
#We filter the dataset to include only Nairobi County
nrb_healthcare_services = df[df['County'] == 'Nairobi']

#Then group by facility and count the occurrences of each owner
owner_counts = nrb_healthcare_services.groupby(['Owner']).size().reset_index(name='Owner Count')

#Sort the owner counts in descending order to identify the owners with the highest count
owner_counts_sorted = owner_counts.sort_values(by='Owner Count', ascending=False)

#We print the owners and their counts for facilities in Nairobi
print("Owners and counts for facilities in Nairobi County:")
print(owner_counts_sorted.head(10))
```

```python
#Set the plot style
sns.set(style="whitegrid")

#Create the plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Owner Count', y='Owner', data=owner_counts_sorted.head(10), palette='viridis')

#Add labels and title
plt.xlabel('Owner Count')
plt.ylabel('Owner')
plt.title('Top 10 Owners of Healthcare Facilities in Nairobi County')

#Show our plot
plt.tight_layout()
plt.show()
```

What are the key challenges and bottlenecks in the healthcare delivery system within Nairobi, such as shortages of medical personnel or inadequate infrastructure?

```python
# We filter the dataset to include only Nairobi County
nrb_healthcare_services = df[df['County'] == 'Nairobi']

# Analyze staffing levels: Count the number of medical personnel per facility
staffing_levels = nrb_healthcare_services.groupby('Facility Name')['Job Title of in Charge'].count()

# Analyze infrastructure adequacy: Check for facilities with zero beds or cots
inadequate_infrastructure = nrb_healthcare_services[(nrb_healthcare_services['Beds'] == 0) | (nrb_healthcare_services['Cots'] == 0)]

# Print the identified challenges and bottlenecks
print("Facilities with shortages of medical personnel:")
print(staffing_levels[staffing_levels == 0])  #For facilities with no medical personnel

print("\nFacilities with inadequate infrastructure (zero beds or cots):")
print(inadequate_infrastructure[['Facility Name', 'Beds', 'Cots']])  #For facilities with inadequate infrastructure
```
